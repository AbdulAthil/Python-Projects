## Introduction

* In this project,we are going to see how to write a python code for the machine which is used by the waiters at the hotels to take orders from the customers and to generate bill for the order.

* Before 10 years,there wasn't so many big hotels in every area and even all those big hotel was not using a computer billing technology.

* But now wherever we are we can fine a good big hotel beacuse the need of hotels becomes high.Everyone wants to eat at hotels these days even if they can go home and eat a homely food.

* And the culture of hand written bill and the suppliers with the notepads in their hands is not we can see at hotels because computers make everyones work much more easier.

* So if you tell the waiter what you want to order,they will update the order in the machine which they have in their hands and the order will be updated automatically in the computer (or) machine in the kitchen area.

* They won't have to run to the kitchen and tell the chef about the order and table number etc.

* And also the bill will be automatically update in the system with the table number the customer sit in.

* For an example,if a customer sit on table number 7 and order a mutton biriyani,the waiter update it in his machine and the order will be updated automatically in the kitchen system and the bill for the order will be updated automatically in the billing system.

* The waiter even can print the bill in his machine and give it to the customer.

* Then the customer get the bill and go to the billing area to pay the bill,he handover the bill to the person who sits behind the computer.He will check the bill with this computer bill which generated automatically when the waiter was updating the order.

* So lots of work can be simplified if the hotel has a computer based order taking and billing technology.

## Requirements of Hotel Food Order

* The available items has to be shown to the waiter so that we could avoid some misunderstandings.

* At first the total available items are stored in the system and it will be reduced based on the orders.

* For an example,if there are 30 meals are available and then 15 of them were sold,the available meals are only 15. It has to be shown in the system for the waiter. So that they can tell the customers if meals are available or not and how many meals are available.

* It is same for all the food items.

* Then the menucard. In menu card it contains the prices of the food items of the hotel. Non veg items,veg items has seperate menu cards and in each menu card,the desert item menu card must be available.

* Non veg and veg can't be ordered at the same time.If the customer needs to order both,the bill will be generated seperatedly which won't be a problem beacuse we can add the two bill values.

* The waiter has to enter the quantity of the items customer ordered in the machine one by one,after main course the desert can be orderd.

* Once the order taken from the customer,there can't be possible to cancel a single or multiple particular items from the order.If a customer wants to cancel a food item they have to cancel the complete order.

* But they can add items as much as they want,but the added items will be on a seperate bill.

* The bill will be generated automatically once the waiter complete the updating process.

* At last,the waiter can be generate the bill from the machine he has on his hands and hand it over to the customer.
