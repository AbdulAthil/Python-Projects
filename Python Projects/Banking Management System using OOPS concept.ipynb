## Requirements of the Banking Management System

* The banking management proccess has to be very secured,so we have to make sure the program is most secured to protect the customer's money and their account from the criminals.
* We have to create an account initially,so that we can make deposit and withdraw money from the accounts we create.

* And we also be able to check the balance in our account and get the account statement of our account.

* These are the basic requirements of the banking management process.

* For create an account,the customer has to provide the basic details such as their name,mobile number,branch location,account type.

* After that they have to set a secret pin number for the account which they don't suppose to tell other persons for the safety purposes.

* The pin number has to be only four digits,they can create any 4 numbers as their account pin number.Account creating process only go to the next step if this condition satisfies.

* The system has to check the details given by the customer along with the previous accounts,so that we can avoid a duplicate account.

* If the account holder name and the pin number is matches with any of the previous accounts at the same time,the account won't be generated.

* If either name or pin number differs from the previous accounts,the account can be generated.

* Every customer has to deposit 5000 rupees or more than 5000 rupees as a initial deposit,if the initial deposit is lesser than 5000,account won't be generated.

* Unique Account number and the customer id are generated automatically for every account.

* And we have to make sure that the system can allow many accounts to be generated for various customers.Each account will have a unique customer id and account number and only the customers are known their account holder name and the pin number which they generated.

* After creating an account,customers has to be allowed to make transactions such as deposit the money in their account and withdraw money from their account.

* Also,they can be able to check their account balance and get their account statement whenever they need.

* They can access their account by entering the pin number and the account holder name,so that even someone find out the pin number for their account,they can't be allowed to enter without the correct account holder name.

* For deposit,they can only deposit upto 50000 at a time. If they want to deposit more than that,they have to make the deposit process multiple times.

* For withdrawn,they can only get how much balance amount they got in their account.
