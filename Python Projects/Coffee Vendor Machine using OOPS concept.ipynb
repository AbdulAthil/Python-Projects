## Introduction

In this project,we are going to write a python code for a coffee machine which could make three different types of coffees such as

## Requirements of the Coffee Machine

* A coffee machine is something which makes various types of coffee as per the user requirements.

* To make a coffee we need water,milk,coffee powder.

* And the user has to pay for the coffee which he want to get.

* Here,we have three types of coffee such as espresso,latte and cappuccino.

* And each one of them required different amount of water,milk and coffee powder. So we have to make sure that each of them had correct compositions.

* And the price for each coffee is also different from one another. So the consumer has to make sure that they pay the required amount of money for the coffee they want to get.

* If the consumer doesn't provide enough money for the coffee they selected,the machine will have to return the money.

* We can only resource the machine with some particular amount of water,milk and coffee powder as a storage.

* For every coffee it makes,the resources has to be updated after completion of the process.If the latte required 200ml of wated and we have 1000ml of water in storage,it has to be deducted from the storage as it makes latte and update the resource in water as 800ml once the process of latte making is completed.

* And if we have only 150ml of water in resources and the consumer wants to get a latte,it won't be possible beacuse there is shortage of water in the resources to make the latte.

* These above two steps goes same for milk and coffee powder.

* The coffee making proccess only begins after all these four requirement (or) conditions got satisfied.

* Otherwise,the machine won't be able to give what the consumer want from it.

* And we need to make sure that the machine can be repired by the workers whenever the need comes.

* For the repairing purposes,the particular employee id numbers has to be entered and use a keyword "off"
